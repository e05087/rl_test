SPY (미국 주식), VSS (글로벌 소형주), TLT (미국 장기 국채)!
*   SPY, VSS 의 1개월, 3개월, 6개월 수익의 평균으로 모멘텀 점수 계산
*   SPY, VSS 중에 모멘텀 점수가 높은 자산을 매수
*   SPY, VSS 모두 모텐덤 점수가 0 이하이면 TLT 매수

자세한 설명:
https://engineeredportfolio.com/2018/05/02/accelerating-dual-momentum-investing/

pct_change : 
https://pandas.pydata.org/docs/reference/api/pandas.Series.pct_change.html

```python
import yfinance as yf
import pandas as pd

df =  yf.download(['SPY', 'VSS','TLT'], start='2010-01-01')['Close']
df1 = df.dropna()
df1 = df1.reset_index()

df1['year'] = pd.to_datetime(df1['Date']).dt.year
df1['month'] = pd.to_datetime(df1['Date']).dt.month
df1['day'] = pd.to_datetime(df1['Date']).dt.day

df2 = df1.sort_values(by=['year','month','day'], ascending=True).drop_duplicates(subset=['year', 'month'], keep='last')
df3 = df2.drop(["year", "month", "day"], axis=1)
df4 = df3.set_index("Date")


mon1 = df4.pct_change(periods=1)
mon3 = df4.pct_change(periods=3)
mon6 = df4.pct_change(periods=6)
mon12 = df4.pct_change(periods=12)

score = mon1+mon3+mon6

print(score.tail())
```

```python
def adm_asset(i):
    asset = 'SPY'
    if score['VSS'].iloc[i]>score['SPY'].iloc[i]:
        asset = 'VSS'
    if score[asset].iloc[i]<0:
        asset = 'TLT'
    return asset  

assets = [''] * 6
for i in range(6, len(score)):
    assets.append(adm_asset(i))

df_stat = pd.DataFrame(columns = ['SPY', 'VSS','TLT'])

for asset in assets:
  stat_info = {'SPY' : False , 'VSS' : False, 'TLT' : False}
  if asset != '':
    stat_info[asset] = True
  df_stat=df_stat.append(stat_info, ignore_index=True)  

df_stat.index = mon12.index
print(df_stat.head(10))
print(df_stat.tail(10))
```

```python
import yfinance as yf
import bt
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

data =  yf.download(['SPY', 'VSS','TLT'], start='2010-01-01')['Adj Close']
data = data.dropna()

RMTEST = bt.Strategy('Accelerating Dual momentum',
 algos = [bt.algos.RunMonthly(run_on_first_date = False, run_on_end_of_period=True, run_on_last_date=True),
 bt.algos.PrintInfo('{name}:{now}. Value:{_value:0.0f}, Price:{_price:0.4f}'),
 bt.algos.SelectAll(),
 bt.algos.SelectWhere(df_stat),
 bt.algos.WeighEqually(),
 bt.algos.PrintTempData(),
 bt.algos.Rebalance()],
)

t = bt.Backtest(RMTEST, data)
report = bt.run(t)

# Plot performance
report.plot(figsize=(15,5))
plt.title('Accelerating Dual momentum Strategy')
plt.show()

# Plot Sector Weights
report.backtests['Accelerating Dual momentum'].security_weights.plot.area(figsize=(15,5))
plt.title("Accelerating Dual momentum Strategy")
plt.legend(ncol=2,loc='right',bbox_to_anchor=(1.2,0.5))
plt.show()
```
